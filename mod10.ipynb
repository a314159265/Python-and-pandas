Use a nested comprehension to build a multiplication table (a list of lists) for 1–5.

```python
mt = [[i * j for j in range(1,6)] for i in range(1,6)]

print(mt)
```

Write a function stats(*numbers) that returns the count, sum, and average of however
many numbers it's given

```python
def stat(*nums) -> list:
    """function taht returns count, sum, and mean of numbers given"""
    total = sum(nums)
    count = len(nums)
    return count, total, total/count if count > 0 else 0

stat(1,2,3)
```

Write a generator that yields the Fibonacci sequence indefinitely, and print the first 10
values from it.

```python
def fg() -> None:
    """function that generate fibonacci sequence"""
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib_gen = fg()
for i in range(10):
    print(next(fib_gen))
```

Given a price like 1234.5 , print it formatted as currency with a thousands separator and
two decimals.

```python
def fm(n:float) -> float:
    """formats price given to it"""
    return f'{n:,.2f}'
    
fm(1234.5)
```
