Task 8 — Filling Missing Values

Fill:

1. Age using the median.
2. Marks using the mean.
3. Department with "Unknown".
4. Name with "Missing".

```python
mf = df.copy()
mf['Age'] = mf['Age'].fillna(mf['Age'].median())
mf['Marks'] = mf['Marks'].fillna(mf['Marks'].mean())
mf['Department'] = mf['Department'].fillna('Unknown')
mf['Name'] = mf['Name'].fillna('Missing')
print(mf)
```

Task 11 — Invalid Values

Find:

1. Students with negative ages.
2. Students whose marks exceed 100.
3. Students whose marks are below 0.
4. Replace negative ages with NaN.
5. Replace marks above 100 with NaN.

```python
# print(df.query("Age < 0"))
# print(df.query("Marks > 100"))
# print(df.query("Marks < 0"))
df = df.replace(np.nan, 0)
df["Age"] = np.where(df["Age"] < 0, np.nan, df["Age"])
df["Marks"] = np.where(df["Marks"] > 100, np.nan, df["Marks"])
print(df)
```
