Task 8 — Filling Missing Values

Fill:

1. Age using the median.
2. Marks using the mean.
3. Department with "Unknown".
4. Name with "Missing".

```python
mf = df.copy()
mf['Age'] = mf['Age'].fillna(mf['Age'].median())
mf['Marks'] = mf['Marks'].fillna(mf['Marks'].mean())
mf['Department'] = mf['Department'].fillna('Unknown')
mf['Name'] = mf['Name'].fillna('Missing')
print(mf)
```

Task 11 — Invalid Values

Find:

1. Students with negative ages.
2. Students whose marks exceed 100.
3. Students whose marks are below 0.
4. Replace negative ages with NaN.
5. Replace marks above 100 with NaN.

```python
# print(df.query("Age < 0"))
# print(df.query("Marks > 100"))
# print(df.query("Marks < 0"))
df = df.replace(np.nan, 0)
df["Age"] = np.where(df["Age"] < 0, np.nan, df["Age"])
df["Marks"] = np.where(df["Marks"] > 100, np.nan, df["Marks"])
print(df)
```

Task 12 — Text Cleaning

Clean the Gender column so every value becomes either:

male
female

Clean the Name column by removing spaces.

Clean the Department column so:

Computer Science
computer science

become

Computer Science

and

Information Technology
IT

become

```python
df["Gender"]=df["Gender"].str.strip().str.lower()
df["Name"]=df["Name"].str.strip()
df["Department"]=np.where(df["Department"].str.strip().str.replace(" ", "").str.lower() == "computerscience", df["Department"].str.title(), df["Department"])
df["Department"]=np.where(df["Department"].str.strip().str.replace(" ", "").str.lower()=="informationtechnology", 'IT', df["Department"])
print(df)
```

Task 13 — Datatypes
1. Display current datatypes.
2. Convert Enrollment Date to datetime.
3. Extract:
Year
Month
Day
Day name

into new columns.

```python
df.info()
df["Enrollment Date"] = pd.to_datetime(df["Enrollment Date"])
df["year"]=df["Enrollment Date"].dt.year
df["Month"]=df["Enrollment Date"].dt.month
df["day"]=df["Enrollment Date"].dt.day
df["day name"]=df["Enrollment Date"].dt.day_name()
print(df)
```
