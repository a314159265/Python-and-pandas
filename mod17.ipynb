```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
ROOT_DIR = Path.cwd()
ss = ROOT_DIR/'data'/'superstoreSales.csv'
df = pd.read_csv(ss, encoding='latin1')
```

Create a band column ( "high" / "low" by whether Sales > 500) two ways — with apply
and with np.where — and confirm they match.

```python
df['certification'] = np.where(df['Sales'] > 500, 'high', 'low')
df['apply_cert'] = df['Sales'].apply(
    lambda x: 'high' if x>500 else 'low'
)
df
```
