Build a wide table of total Sales with Region as rows and Product Category as columns

```python
import pandas as pd
from pathlib import Path
```

```python
ROOT_DIR = Path.cwd()
ss = ROOT_DIR/'data'/'superstoreSales.csv'
df = pd.read_csv(ss, encoding='latin1')
```

```python
dw = df.pivot_table(index='Region', columns='Product Category', values='Sales', aggfunc='sum')
dw
```

```python
dm = pd.melt(df, id_vars=['Region'], var_name='Product Category', value_name='Sales1')
dm
```

```python
dww = dm.pivot_table(index='Region', columns='Product Category', values='Sales', aggfunc='sum')
dww
```

```python
# Flatten the wide table back to long format
dm_long = pd.melt(
    dww.reset_index(), 
    id_vars=['Region'], 
    var_name='Product Category', 
    value_name='Sales'
)
dm_long
```
